```python
def my_func():
    return {'res':1}

node1 =  PyNode(f"a", my_func,output_vars={'res'})
node2 =  PyNode(f"b", my_func,output_vars={'res'},input_vars={'in':node1.outputs['res']})
node3 =  PyNode(f"c", my_func,output_vars={'res'},input_vars={'in':node2.outputs['res']})
node4 =  PyNode(f"d", my_func,output_vars={'res'},must_run_after={node2})

experiment = Graph(None, {node1,node2,node3,node4}, "run_a") #No resources used here
handle = experiment.run()
graph = handle.dot_graph()
graph.graph_attr['rankdir'] = 'LR'
graph.format='svg'
# graph.save('this.png')
graph.render('DAG')
```
